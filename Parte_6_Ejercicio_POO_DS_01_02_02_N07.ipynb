## Ejercicio 1
Somos una empresa que tiene varias tiendas de electrodomesticos. Necesitamos un programa para manejar las tiendas, ver las ventas que han tenido, cuántos empleados hay, etc... Para ello vamos a modelizar la tienda en una clase, que tendrá las siguientes características:
* Nombre clase: "Tienda"
* Atributos comunes:
    * Tipo: "Electrodomésticos"
    * Abierta: True
* Atributos propios de cada tienda:
    * Nombre - String
    * Dirección - String
    * Número de empleados - int
    * Ventas ultimos 3 meses - Lista de 3 numeros
* Método para calcular las ventas de todos los meses, que devuelva un numero con todas las ventas.
* Método que calcula la media de ventas de los ultimos meses, por empleado
* Método que devuelve en un string el nombre de la tienda, junto con su dirección.
* Método que obtiene las ventas del último mes.
* Método para dar una proyección de las ventas en caso de haber invertido X dinero en marketing en esa tienda. Siendo X el parámetro de entrada. Si X es menor de 1000, las ventas de los ultimos 3 meses hubiesen sido de (1.2 x ventas), si es mayor o igual de 1000, las ventas hubiesen sido de (1.5 x venta). El método reescribe el atributo de ventas en los últimos 3 meses, y además devuelve las ventas con el aumento.

Se pide:
1. Implementar la clase "Tienda"
2. Crear tres tiendas con datos inventados
3. Comprueba en al menos una de ellas, todo lo implementado en la clase tienda (sus atributos, media de ventas, ventas/empleado...)
3. Calcular las ventas del último mes de todas las tiendas. Para ello usa el bucle `for`
4. Imprime por pantalla los nombres de las tiendas cuya dirección lleve el string "Avenida"
5. Seguro que ya has hecho este apartado... Documenta la clase :)

```python
class Tienda:

    """
    Representa una tienda de electrodomésticos.

    Atributos:
        nombre (str): Nombre de la tienda.
        direccion (str): Dirección de la tienda.
        num_empleados (int): Número de empleados.
        ventas_3_meses (list): Lista con las ventas de los últimos 3 meses.
        ventas_proyectadas (list): Ventas proyectadas después de la inversión.

    Métodos:
        ventas_totales(): Retorna la suma de las ventas de los últimos 3 meses.
        media_ventas(): Calcula el promedio de ventas por empleado.
        ubicacion(): Retorna una cadena con el nombre y la dirección de la tienda.
        ultimo_mes(): Retorna las ventas del último mes.
        inversion(x): Aplica un multiplicador a las ventas basado en la inversión x.
    """
    
    tipo = "Electrodomesticos"
    abierta = True
    def __init__(self, nombre, direccion, num_empleados, ventas_3_meses):


        """
        Inicializa los atributos de la tienda.

        Args:
            nombre (str): Nombre de la tienda.
            direccion (str): Dirección de la tienda.
            num_empleados (int): Número de empleados.
            ventas_3_meses (list): Ventas de los últimos 3 meses.
        """
        self.nombre = str(nombre)
        self.direccion = str(direccion)
        self.num_empleados = int(num_empleados)
        self.ventas_3_meses = ventas_3_meses
        self.ventas_proyectadas = []
    def ventas_totales(self):
        """
        Calcula la suma total de las ventas de los últimos 3 meses.

        Returns:
            float: La suma total de ventas.
        """
        ventas_totales = sum(self.ventas_3_meses)
        return ventas_totales
    def media_ventas(self):
        """
        Calcula la media de ventas por empleado.

        Returns:
            float: Media de ventas por empleado.
        """
        media_ventas = self.ventas_totales() / self.num_empleados
        return media_ventas
    def ubicacion(self):
        """
        Retorna una cadena con el nombre y dirección de la tienda.

        Returns:
            str: Información de ubicación.
        """
        ubicacion = "Nombre de la tienda: " + self.nombre + " y nombre de la direccion: " + self.direccion
        return ubicacion
    def ultimo_mes(self):
        """
        Obtiene las ventas del último mes.

        Returns:
            float: Ventas del último mes.
        """
        ultimo_mes = self.ventas_3_meses[-1]
        return ultimo_mes
    def inversion (self, x):
        """
        Aplica un multiplicador a las ventas basado en el monto de la inversión.

        Args:
            x (float): Monto invertido.

        Returns:
            float: Suma de las ventas proyectadas después de la inversión.
        """
        if x < 1000:
            multiplicador = 1.2
        else:
            multiplicador = 1.5
        nuevas_ventas = []
        for ventas_mes in self.ventas_3_meses:
            z = multiplicador * ventas_mes
            nuevas_ventas.append(z)
        self.ventas_proyectadas = nuevas_ventas
        return sum(self.ventas_proyectadas)
        



        
tienda1 = Tienda("Madrid Centro", "Calle Gran Vía, Madrid", 10, [1500, 1800, 1700])
tienda2 = Tienda("El Rastro", "Plaza Mayor, Madrid", 7, [900, 1100, 950])
tienda3 = Tienda("Avenida Sol", "Avenida de América, Madrid", 12, [1300, 1400, 1350])


print(tienda1.nombre)
print(tienda1.direccion)
print(tienda1.num_empleados)
print(tienda1.ventas_3_meses)
print(tienda1.ventas_totales())
print(tienda1.media_ventas())
print(tienda1.ubicacion())
print(tienda1.ultimo_mes())
print(tienda1.inversion (1500))




lista_de_tiendas = [tienda1, tienda2, tienda3]
for tienda in lista_de_tiendas:
    if "Avenida" in tienda.direccion:
        print(tienda.nombre)
```

## Ejercicio 2
En este ejercicio vamos a implementar una clase *Perro* en Python. La clase tiene las siguientes características:
* Cosas que sabemos seguro que tiene un perro
    * Tiene 4 patas
    * 2 orejas
    * 2 ojos
    * Una velocidad de 0. Por defecto, el perro está parado
* Cuando se inicialice:
    * El perro será de una determinada raza
    * Por defecto tendrá pelo "Marrón", a no ser que se diga lo contrario.
    * Por defecto no tendrá dueño, a no ser que se diga lo contrario.
    
* Dispondrá también de un método llamado andar, que tiene un argumento de entrada (aumento_velocidad). Este valor se le sumará a la velocidad que ya llevaba el perro.
* Necesita otro método (parar), donde pondremos la velocidad a 0.
* Otro método llamado "ladrar", que tendrá un argumento de entrada, y la salida será el siguiente string: "GUAU!" + el argumento de entrada.


Se pide:
* Implementa la clase *Perro*
* Crea un objeto de tipo *Perro*, sin dueño
* Comprueba que están bien todos sus atributos
* Prueba que ande, y comprueba su velocidad
* Páralo
* Documenta la clase *Perro*

```python
class Perro:
    """
    Cosas que sabemos seguro que tiene un perro

    Atributos: 
        Patas
        Orejas
        Ojos
        Velocidad
        Raza
        Pelo
        Dueno
    Metodos:
        Andar
        Parar
        Ladrar
    
    """
    patas = 4
    orejas = 2
    ojos = 2
    velocidad = 0
    
    def __init__(self, raza, pelo="Marron", dueno=None):
        """
        Inicializa un nuevo objeto Perro.
        """

        self.raza = raza
        self.pelo = pelo
        self.dueno = dueno
        
    def andar(self, aumento_velocidad): 
        """
        Aumenta a velocidad del perro.
        
        """
        self.velocidad += aumento_velocidad
        
    def parar(self):
        """
        Detiene al perro poniendo la velocidad en 0.
        """
        self.velocidad = 0

    def ladrar(self, hablar):
        """
        Genera un ladrido con texto.
        """
        hablar = "GUAU! " + hablar
        return hablar
         
        
        
perro = Perro("Pastor alemán")
print(perro.raza)
print(perro.pelo)
print(perro.dueno)
print(perro.velocidad)
print(perro.patas)
print(perro.orejas)
print(perro.ojos)


perro.andar(10)
print(perro.velocidad)
perro.parar()
print(perro.velocidad)
```
